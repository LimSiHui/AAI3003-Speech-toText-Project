```python
import re
import matplotlib.pyplot as plt

# Assuming the data is stored in a file named 'wer_data.txt'
file_path = '/content/testing2_wer_score (30 epochs).txt'

# Function to read the data from a file and extract epoch numbers and WER values
def read_wer_data(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
    pattern = r"Epoch (\d+), PRED_\d+: WER = (\d+\.\d+)"
    return re.findall(pattern, data)

# Function to calculate the average WER for each epoch
def calculate_average_wer(matches):
    wer_data = {}
    for epoch, wer in matches:
        epoch = int(epoch)
        wer = float(wer)
        wer_data.setdefault(epoch, []).append(wer)
    return {epoch: sum(wers) / len(wers) for epoch, wers in wer_data.items()}

# Function to plot the average WER for each epoch
def plot_average_wer(average_wer):
    plt.figure(figsize=(8, 5))
    plt.plot(average_wer.keys(), average_wer.values(), marker='o')
    plt.title('Hubert - Average WER per Epoch (30 epochs)')
    plt.xlabel('Epoch')
    plt.ylabel('Average WER')
    plt.grid(True)
    plt.show()

# Read, calculate, and plot the WER data
matches = read_wer_data(file_path)
average_wer = calculate_average_wer(matches)
plot_average_wer(average_wer)

```

```python
# Assuming the data provided by the user is stored in a text file, let's simulate reading from a text file
# First, we will create a temporary text file with the data

file_path = '/content/deep_speech_wer_score.txt'

# Now, let's read from this file and plot the graph
epochs = []
wers = []

# Read the file and process the data
with open(file_path, 'r') as file:
    for line in file:
        epoch_str, wer_str = line.strip().split(': ')
        epochs.append(int(epoch_str.split(' ')[1]))
        wers.append(float(wer_str))

# Plot the data
plt.figure(figsize=(8, 5))
plt.plot(epochs, wers, marker='o', linestyle='-')
plt.title('DeepSpeech - WER per Epoch (50 epochs)')
plt.xlabel('Epoch')
plt.ylabel('WER')
plt.grid(True)
plt.show()
```
